***Project Freight***

1) What is the project freight?
2) Most of the project freight do not have quantities, nor do they have part descriptions
3) 70 of the project freight has part descriptions and associated quantities, whats the difference? 
4) How are these purchase orders any different, are they linked to a supplier or is it location dependent? (its not a supplier thing, 27 suppliers are common in both null and not null freight) 

```python
freight_use = data.loc[data['ACCOUNT DESCRIPTION'] == 'PROJECT Freight', ['PART DESCRIPTION', 'SUPPLIER NAME','PO INV QTY']]
print(freight_use)
```

```python
null_freight = freight_use.loc[freight_use['PO INV QTY'].isna(), ['PART DESCRIPTION', 'SUPPLIER NAME']]
print(null_freight)
```

```python
not_null_freight= freight_use.loc[freight_use['PO INV QTY'].notna() & (freight_use['PO INV QTY'] != ''),]
print (not_null_freight) 
```

```python
# Check is its a supplier thing 
null_suppliers = null_freight['SUPPLIER NAME'].dropna().unique()
not_null_suppliers = not_null_freight['SUPPLIER NAME'].dropna().unique()

# Find the intersection of both supplier name sets
common_suppliers = set(null_suppliers).intersection(set(not_null_suppliers))

print(len(common_suppliers))
```
