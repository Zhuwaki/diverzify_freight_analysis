```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = pd.read_csv('../../data/downloads/freight_comparison.csv')
df = pd.DataFrame(data)

# Sort for smoothing
df_sorted = df.sort_values("pct_priority")
#df_sorted["smoothed"] = df_sorted["pct_freight"].rolling(window=5, center=True, min_periods=1).mean()

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df["pct_priority"], df["pct_freight"], color="gray", label="Invoice Data")
plt.plot(df_sorted["percentage_priority"], df_sorted["smoothed"], color="blue", linewidth=2, label="Trend (Smoothed)")

# Annotations
plt.axhline(6, color="red", linestyle="--", linewidth=1)
plt.axvline(10, color="orange", linestyle="--", linewidth=1)
plt.text(2, 9, "⚠️ Inefficient Zone", color="red", fontsize=9)

plt.axhline(5, color="green", linestyle="--", linewidth=1)
plt.axvline(95, color="green", linestyle="--", linewidth=1)
plt.text(70, 2, "✅ Efficient Zone", color="green", fontsize=9)

plt.title("Freight Cost vs Priority Commodities on Invoice", fontsize=14)
plt.xlabel("Priority Commodities (%)")
plt.ylabel("Freight Share (%)")
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```
