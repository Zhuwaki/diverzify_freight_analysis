```python
# Load vnl dataset
file_path = "../../data/input/1VNL_data.xlsx"
df_vnl = pd.read_excel(file_path, sheet_name="1VNL (LVT, LVP, VCT)", engine="openpyxl")

# Load dana1 dataset
file_path = "../../data/input/AF Commercial Data_100324.xlsx"
df_af = pd.read_excel(file_path, sheet_name="Item", engine="openpyxl")

# Load dana2 dataset
file_path = "../../data/input/AHF Contract Data_100424.xlsx"
df_ahf = pd.read_excel(file_path, sheet_name="Item", engine="openpyxl")

# Load dana3 dataset LVT Collections
file_path = "../../data/input/Mannington LVT Master List_Latest Version.xlsx"
df_man = pd.read_excel(file_path, sheet_name="LVT Collections", engine="openpyxl")
```

```python
#emove the range and bracket from the column names
df_vnl.columns = df_vnl.columns.str.replace(r"Range\[|\]", "", regex=True)
df_vnl.head(2)
```

```python
# Count the number of instances where part_no from df_vnl exists in df_af['ItemNumber']
matching_count = df_vnl['part_no'].isin(df_af['ItemNumber']).sum()
print(f"Number of matching part numbers: {matching_count}")
```

```python
matching_count_ahf = df_vnl['part_no'].isin(df_ahf['ItemNumber']).sum()
print(f"Number of matching part numbers in df_ahf: {matching_count_ahf}")
```

```python
# Find part numbers in df_vnl where part_description contains ItemName_en from df_ahf
matching_parts = df_vnl[df_vnl['part_description'].str.contains('|'.join(df_ahf['ItemName_en']), case=False, na=False)]

# Display the matching part numbers
print(matching_parts[['part_no', 'part_description']])
```

```python
# Find part numbers in df_vnl where part_description contains ItemName_en from df_af
matching_parts_af = df_vnl[df_vnl['part_description'].str.contains('|'.join(df_af['ItemName_en']), case=False, na=False)]

# Display the matching part numbers
print(matching_parts_af[['part_no', 'part_description']])
```
